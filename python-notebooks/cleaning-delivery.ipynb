```python
merged_df = merged_df.drop_duplicates(subset='order_id', keep='first')
```

```python
merged_df['accept_gps_lng']=merged_df['accept_gps_lng'].fillna(merged_df['accept_gps_lng'].mean())
merged_df['accept_gps_lat']=merged_df['accept_gps_lat'].fillna(merged_df['accept_gps_lat'].mean())
```

```python
merged_df.isna().any()
```

```python
# Converting accept_time and delivery_time to datetime
merged_df['accept_time'] = pd.to_datetime(merged_df['accept_time'], format='%m-%d %H:%M:%S', errors='coerce')
merged_df['delivery_time'] = pd.to_datetime(merged_df['delivery_time'], format='%m-%d %H:%M:%S', errors='coerce')
merged_df['accept_gps_time'] = pd.to_datetime(merged_df['accept_gps_time'], format='%m-%d %H:%M:%S', errors='coerce')
merged_df['delivery_gps_time'] = pd.to_datetime(merged_df['delivery_gps_time'], format='%m-%d %H:%M:%S', errors='coerce')
```

```python
merged_df
```

```python
# Ensuring lng and lat are float
merged_df['lng'] = merged_df['lng'].astype(float)
merged_df['lat'] = merged_df['lat'].astype(float)

# Converting aoi_type to categorical
merged_df['aoi_type'] = merged_df['aoi_type'].astype('category')

# Verifying changes
merged_df.dtypes
```

```python
# Calculating delivery duration in minutes
merged_df['delivery_duration'] = (merged_df['delivery_time'] - merged_df['accept_time']).dt.total_seconds() / 60

# Removing rows with missing or invalid durations
merged_df = merged_df.dropna(subset=['delivery_duration'])

# Calculating IQR for delivery duration
Q1 = merged_df['delivery_duration'].quantile(0.25)
Q3 = merged_df['delivery_duration'].quantile(0.75)
IQR = Q3 - Q1

# Identifying lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtering out outliers
filtered_data = merged_df[(merged_df['delivery_duration'] >= lower_bound) & (merged_df['delivery_duration'] <= upper_bound)]

# Verifying the results
filtered_data.shape, merged_df.shape, lower_bound, upper_bound
```

```python
filtered_data
```

```python
# Function to validate latitude and longitude ranges
def validate_coordinates(df, lat_column, lng_column):
    invalid_lat = df[(df[lat_column] < -90) | (df[lat_column] > 90)]
    invalid_lng = df[(df[lng_column] < -180) | (df[lng_column] > 180)]
    return invalid_lat, invalid_lng

# Validate the columns
invalid_primary_lat, invalid_primary_lng = validate_coordinates(merged_df, 'lat', 'lng')
invalid_accept_lat, invalid_accept_lng = validate_coordinates(merged_df, 'accept_gps_lat', 'accept_gps_lng')
invalid_delivery_lat, invalid_delivery_lng = validate_coordinates(merged_df, 'delivery_gps_lat', 'delivery_gps_lng')

# Combine results to identify issues across all columns
invalid_entries = pd.concat([
    invalid_primary_lat, invalid_primary_lng,
    invalid_accept_lat, invalid_accept_lng,
    invalid_delivery_lat, invalid_delivery_lng
]).drop_duplicates()

invalid_entries.shape, invalid_entries
```
